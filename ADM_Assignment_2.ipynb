```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("employee_performance.csv")
print("Original Dataset:\n", df.head())
```

```python
print("\nStatistical Summary:\n", df.describe())
print("\nDataset Info:\n")
df.info()
```

```python
df['Department_Encoded'] = df['Department'].astype('category').cat.codes
```

```python
df['Bonus'] = df['Monthly_Salary'] * 0.1
```

```python
performance_summary = df.groupby('Department')['Performance_Score'].mean().reset_index()
print("\nAverage Performance by Department:\n", performance_summary)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Monthly_Salary'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Monthly Salaries')
plt.xlabel('Monthly Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Years_of_Experience'], df['Performance_Score'], color='purple')
plt.title('Years of Experience vs. Performance Score')
plt.xlabel('Years of Experience')
plt.ylabel('Performance Score')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
correlation_matrix = df[['Monthly_Salary', 'Years_of_Experience', 'Performance_Score']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Numerical Features')
plt.show()
```

```python
df.to_csv("processed_employee_performance.csv", index=False)
print("\nProcessed dataset saved as 'processed_employee_performance.csv'")
```
